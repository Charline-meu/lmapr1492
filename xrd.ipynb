# Tâche 7: Calcul de pics d'un diffractogramme

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
# télécharger le ﬁchier CIF du matériau qui vous a été attribué

with MPRester("msKcAm8eGLQeLLbWk7xJ") as m:
    # !!! je me suis connectée sur l'ancien site de materials project car l'autre API key ne fonctionnait pas bien !!!    
    structure = m.get_structure_by_material_id("mp-961687")
    structure.to(filename = "ZrSnPd.cif")
```

```python
# Récupération des informations concernant le diffractogramme

xrd = XRDCalculator("CuKa")
pat = xrd.get_pattern(structure, scaled=True)
```

## Calcul des 3 premiers pics du diffractogramme

```python
for i in range (3):
    if (i == 0):
        print("\033[1m Premier pic pour un rayonnement X incident CuKα de longueur d’onde: λ=1.54060 Å \033[0m")
    else: 
        print("\033[1m",i+1,"ème pic pour un rayonnement X incident CuKα de longueur d’onde: λ=1.54060 Å \033[0m")
    print("Indices hkl et multiplicité:", pat.hkls[0])
    print("Intensité du pic:", pat.y[0])
    print("Angle 2\u03B8(°):", pat.x[i],"°")
    print("Distance associée:", pat.d_hkls[i])
    print("\n")
```

## Diffractogramme

```python
print("Intensité des trois premiers pics en fonction des angles pour un rayonnement X incident CuKα de longueur d’onde: λ=1.54060 Å")
xrd.show_plot(structure,two_theta_range=(0,45))
```
